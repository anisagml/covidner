```python
new_search = "#covid+indonesia -filter:retweets"

tweets = tw.Cursor(api.search, 
                           q=new_search,
                           lang="id",
                           since=date_since, tweet_mode='extended').items(1000)

users_locs = [[tweet.user.id,tweet.user.name, tweet.user.screen_name, tweet.user.location, tweet.id ,tweet.full_text, tweet.created_at] for tweet in tweets]
tweet = pd.DataFrame(data=users_locs, 
                    columns=['user_id','name','user', "location",'id','text','date'])
tweet.to_csv("output1.csv")
tweet
```

```python
new_search = "#covid id -filter:retweets"

tweets = tw.Cursor(api.search, 
                           q=new_search,
                           lang="id",
                           since=date_since, tweet_mode='extended').items(1000)

users_locs = [[tweet.user.id,tweet.user.name, tweet.user.screen_name, tweet.user.location, tweet.id ,tweet.full_text, tweet.created_at] for tweet in tweets]
tweet2 = pd.DataFrame(data=users_locs, 
                    columns=['user_id','name','user', "location",'id','text','date'])
tweet2.to_csv("output2.csv")
tweet2
```

```python
covid = pd.merge(tweet, tweet2, how = 'outer', indicator = True)
covid
```
